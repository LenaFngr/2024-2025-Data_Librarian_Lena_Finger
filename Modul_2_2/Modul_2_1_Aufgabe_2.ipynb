## Datenaufbereitung

```python
import json
import pandas as pd
```

```python
# Daten aus Aufgabe_1 laden
data = json.load(open("Daten_Aufgabe_1.json"))
```

```python
type(data) # Liste muss in DataFrame umgewandelt werden
```

```python
df = pd.DataFrame(data)
df
```

### Daten filtern: Eintr√§ge des Typs "Publication"

```python
df_pubs = df.loc[df["type"]== "Publication"]
# alternativ: df_pubs = df.loc[df.type == "Publication"]
```

### Daten (Spalten) ausw√§hlen und umbenennen

```python
df_pubs.columns
```

```python
df_selection = df_pubs[["type", "id", "tags", "label", "subtitle", "description", "abstract", "author", "authors", "year", "pages", "journal", "publisher", "url", "isbn", "issn", "doi", "sprache", "language", "price", "keyword"]]
```

```python
df_selection = df_selection.rename(columns={"label": "title"})
```

```python
df_selection
```

### Duplikate entfernen

```python
# Entfernen von Duplikaten in der Spalte "title"
# keep = "first" beh√§lt jeweils die erste von doppelten Zeilen
df_selection = df_selection.drop_duplicates(subset = "title", keep = "first") # von 3526 auf 3225 Eintr√§ge reduziert
```

### Daten s√§ubern: Eintr√§ge zu Titel, Autor und Jahr √ºberpr√ºfen

```python
# pr√ºfen, ob in Spalte Titel, Autor und Jahr jede Zeile einen Eintrag hat
print(df_selection["title"].isna().sum()) # Anzahl der Zeilen ohne Titel liegt bei 0
```

```python
print(df_selection["author"].isna().sum()) # Anzahl der Zeilen ohne Autor liegt bei 273
```

```python
print(df_selection["year"].isna().sum()) # Anzahl der Zeilen ohne Jahr liegt bei 0
```

```python
# Daten ausw√§hlen, wo in Spalte "author" ein Wert eingetragen ist
df_selection = df_selection.loc[df_selection["author"].isnull() == False]
```

#### Korrektur der Spalte "year"

```python
# .unique() gibt alle Werte eines Series Objektes aus
df_selection["year"].unique()
# Spalte enth√§lt nicht nur ganze Jahreszahlen
```

```python
# √úber .apply kann eine Funktion auf jeden Wert einer Spalte angewendet werden
# "axis=1" beschreibt, dass Zeilenweise vorgegangen werden soll
# "lambda row:" beschreibt, dass row hier jede iteration, in diesem Fall jede Zeile beschreibt

# Wir schreiben eine neue Spalte "year_correct", in der √ºber row.year.isdigit() √ºberpr√ºft wird,
# ob es der eintrag in "year" Zahl ist bzw. in eine Ganzzahl umgewandelt werden kann
df_selection["year_correct"] = df_selection.apply(lambda row: row.year.isdigit(), axis=1)

# die auftretende warnung kann an dieser Stelle ignoriert werden
```

```python
# es sollen nur zeile im dataframe bleiben mit einem echten jahreseintrag
df_selection_year = df_selection.loc[df_selection.year_correct == True]
```

```python
df_selection_year
```

```python
df_selection_year["year"].unique()
# nur noch ganze Zahlen als Jahresangabe
# eine Zeile mit dem Wert 20070901 muss noch entfernt werden
```

```python
# Zahl der Spalte year in einen integer (ganzzahlenwert) umwandeln und in neuer Spalte speichern
df_selection_year["year_number"] = df_selection_year["year"].astype(int)
```

```python
# entfernen von falschem Wert, in dem alle Werte √ºber 2025 hinaus entfernt werden
df_selection_year_1 = df_selection_year.loc[df_selection_year.year_number <= 2025]
# Problem: andere Tippfehler zb nur 2 oder 3 Ziffern werden nicht erfasst

# m√∂gliche L√∂sung: Daten filtern mit Jahreszahlen zwischen 1000 und 9999
df_selection_year_2 = df_selection_year.loc[df_selection_year["year_number"].between(1000, 9999)]
# oder: string-L√§nge von 4 √ºberpr√ºfen
df_selection_year_3 = df_selection_year.loc[df_selection_year["year_number"].astype(str).str.len() == 4]
```

```python
df_selection_year_3.groupby("year_number").count().title
# da jede Publikation einen titel hat wird das vorkommen der Titel gez√§hlt
```

```python
# neue Spalte bestht aus integer-Werten
df_selection_year_3.info()
```

```python
# nun kann auch bspw. nach Jahreszahl 2010 und aufsteigend gefiltert werden
df_selection_year_3.loc[df_selection_year_3.year_number >= 2010] # 1184 Eintr√§ge 
```

```python
# l√∂schen von ungenutzten spalten
df_selection_year_3 = df_selection_year_3.drop(columns=['year', 'year_correct'])
```

```python
# √ºberpr√ºfen der Anzahl von Autoren und abspeichern in neuer Spalte
df_selection_year_3["number_authors"] = df_selection_year_3.apply(lambda row:len(row.authors), axis = 1)
```
