```python
# packages laden
import pyterrier as pt
import pickle
import nltk
import numpy as np
```

```python
# Pyterrier starten
if not pt.started():
    pt.init()
```

```python
# Felder bzw. Spalten docno und text notwendig, damit pyterrier funktioniert
df['docno'] = df['id']
df['text'] = df['title']
```

```python
df
```

```python
# für Indexierung Ordnerstruktur anlegen
index_folder = "./df_unternehmen_index"

indexer = pt.DFIndexer(index_folder, overwrite = True)
index_ref = indexer.index(df['text'], df['docno'])

print(f"Pfad zum Index: {index_ref.toString()}")
```

```python
index = pt.IndexFactory.of(index_ref)
```

```python
# Anzeige, wie viele Terme in den Dokumenten extrahiert wurden: 4290
print(index.getCollectionStatistics().toString())
```

```python
# Anzeige aller Terme im Korpus und deren Häufigkeit
# Bspw. kommt der Term "extern" in insgesamt 27 Dokumenten vor
# dabei: NT = Menge der Dokumente, die deTerm beinhalten, TF = absolute Häufigkeit des Terms 

for i in index.getLexicon():
    print(f"{i.getKey()} -> {i.getValue().toString()}")
```

```python
# Anzeige der Häufigkeit eines bestimmten Terms 
term = "extern"
index.getLexicon()[term].toString()
```

```python
# Häufigkeit der Terme betrachten
term_freq = {}
for i in index.getLexicon():
    term_freq[i.getKey()] = i.getValue().frequency

# Häufigkeit sortieren (absteigend)
term_freq_abst = sorted(term_freq.items(), key=lambda x: x[1], reverse=True)
```

```python
# Anzeige der 30 häufigsten Terme
term_freq_abst[:29]
```

```python
# Funktion für Berechnung der inversen Dokumente Häufigkeit für einen term
def get_idf(term, index):
    lex = index.getLexicon()
    stemmed_term = stemmer.stem(term)
    if not stemmed_term in lex:
        return
    lex_entry = lex[stemmed_term]
    df_term = lex_entry.getDocumentFrequency()
    N = index.getCollectionStatistics().numberOfDocuments

    idf = N/df_term

    # log10 auf IDF anwenden
    log_idf = np.emath.logn(10, idf)
    return log_idf
```
