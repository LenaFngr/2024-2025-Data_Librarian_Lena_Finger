## Datenquellen suchen

```python
import requests
import json
```

```python
# Daten zu Suchbegriff "Unternehmen"
url = "https://www.bibsonomy.org/json/search/unternehmen?items=1000&duplicates=merged"
result = requests.get(url)
data_unt = result.json() #response Obejkt bietet diese Mehode an
type (data_unt)
```

```python
print(json.dumps(data_unt, indent = 2)) # besser lesbare Ausgabe
```

```python
print(data_unt.keys()) #Anzeige der vorhandenen Keys
```

```python
# Daten zu "Open Innovation"
url = "https://www.bibsonomy.org/json/search/open%20innovation?items=1000"
result = requests.get(url)
data_oi = result.json() #response Obejkt bietet diese Mehode an
type (data_oi)
```

```python
print(data_oi.keys())
```

```python
# Daten zu R&D
url = "https://www.bibsonomy.org/json/search/r&d?items=1000&duplicates=merged"
result = requests.get(url)
data_rnd = result.json() #response Obejkt bietet diese Mehode an
type (data_rnd)
```

```python
print(data_rnd.keys())
```

```python
# Daten zu Unternehmenskultur
url = "https://www.bibsonomy.org/json/search/unternehmenskultur?items=1000"
result = requests.get(url)
data_kult = result.json() #response Obejkt bietet diese Mehode an
type (data_kult)
```

```python
print(data_kult.keys())
```

## Daten speichern

```python
with open('data_unternehmen.json', 'w') as outfile:
    json.dump(data_unt, outfile, indent=2)
```

```python
with open('data_openinnovation.json', 'w') as outfile:
    json.dump(data_oi, outfile, indent=2)
```

```python
with open('data_rnd.json', 'w') as outfile:
    json.dump(data_rnd, outfile, indent=2)
```

```python
with open('data_unternehmenskultur.json', 'w') as outfile:
    json.dump(data_kult, outfile, indent=2)
```

## Daten bearbeiten und zusammenfhren

```python
import pandas as pd
```

```python
type(data_unt["items"]) # Liste von dictionaries, mit Pandas als Dataframe einlesen
```

```python
df_unt = pd.DataFrame(data_unt["items"])
```

```python
# abgespeicherte JSON Dateien laden

# JSON Dateien wurden hier noch einmal neu eingelesen, da fehler bei erster durchfhrung mit den keywords
# neues einlesen ist nicht notwendig
# es kann genauso mit df_unt etc weiter gearbeitet werden statt df_1

data1 = json.load(open("data_unternehmen.json"))
data2 = json.load(open("data_openinnovation.json"))
data3 = json.load(open("data_rnd.json"))
data4 = json.load(open("data_unternehmenskultur.json"))
```

```python
# DataFrames erstellen
df_1 = pd.DataFrame(data1["items"])
df_2 = pd.DataFrame(data2["items"])
df_3 = pd.DataFrame(data3["items"])
df_4 = pd.DataFrame(data4["items"])
```

```python
df_1["keyword"] = "Unternehmen"
df_2["keyword"] = "OI"
df_3["keyword"] = "RnD"
df_4["keyword"] = "Unternehmenskultur"
```

```python
df_3  # Keywords zu df_2 und df_3 werden auf den ersten Blick nicht hinzugefgt
# da keyword nicht am ende erscheint wie bei df 1 und df4 
# mit df_3.columns und df_3.keyword kann geprft werden, ob
# 1) splate keyword im Data Frame enthalten ist und 
# 2) welchen Inhalte die Spalte hat 
```

```python
df_3.columns
```

```python
df_3.keyword
```

### Data Frames ber .concat miteinander kombinieren

```python
df = pd.concat([df_1, df_2, df_3, df_4])
```

```python
df # df enthlt alle Zeilen der 4 DataFrames
```

```python
df.keyword
```
